# Early Stoping in NN

```python
x, y = make_moons(n_samples=100, noise=0.1, random_state=101)
```

```python
plt.scatter(x[:,0], x[:,1], c=y,s=100)
```

```python
x_train, x_test , y_train, y_test = train_test_split(x,y,random_state=125)
```

```python
model2 = Sequential()
model2.add(Dense(10, activation='relu', input_dim=2))
model2.add(Dense(10, activation='relu'))
model2.add(Dense(10, activation='relu'))
model2.add(Dense(10, activation='relu'))
model2.add(Dense(10, activation='relu'))
model2.add(Dense(10, activation='relu'))
model2.add(Dense(10, activation='relu'))
model2.add(Dense(10, activation='relu'))
model2.add(Dense(10, activation='relu'))
model2.add(Dense(10, activation='relu'))
model2.add(Dense(10, activation='relu'))
model2.add(Dense(10, activation='relu'))
model2.add(Dense(1, activation='sigmoid'))
model2.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
```

```python
history = model2.fit(x_train, y_train, validation_data=(x_test, y_test),epochs=500)
```

```python
plt.plot(history.history['loss'], label='train')
plt.plot(history.history['val_loss'], label='test')
plt.legend()
plt.show()
```

```python
from tensorflow.keras.callbacks import EarlyStopping
```

```python
callback = EarlyStopping(
    monitor="val_loss",
    min_delta=0.00001,
    patience=15,
    verbose=1,
    mode="auto",
    baseline=None,
    restore_best_weights=False
   
)
```

```python
history = model2.fit(x_train, y_train, validation_data=(x_test, y_test),epochs=3000, callbacks=callback)
```

```python
plt.plot(history.history['loss'], label='train')
plt.plot(history.history['val_loss'], label='test')
plt.legend()
plt.show()
```
